```python
#En esta parte, se importa la libreria pandas de python, a la cual se llamará pd.
import pandas as pd
```

```python
#En esta parte, se lee el conjunto de datos, especificando tanto el lugar de donde esta almacenado como el nombre de la Hoja.
df = pd.read_excel("/content/Conjunto de datos Accidente de transito Numero Pequeño.xlsx", sheet_name='default_1')
#df = pd.read_excel("/content/Conjunto de datos normalizado MaxMin.xlsx", sheet_name='default_1')
df
```

```python
#En esta parte, el conjunto de datos, llamado df, es copiado a otra variable llamada x, donde en esta ultima
#se realizaran las ultimas labores de limpieza, quitando las columnas string que estaban presentes en el conjunto de datos
#y a su vez, eliminar la columna que se va a ocupar en el siguiente proceso.
x = df.copy()
x = x.drop(['CausasNumber','Idaccidente' ,'Fecha', 'Hora', 'Mes', 'Comuna', 'Región', 'Calidad', 'Sexo', 'Resultado',
            'Accdtes.', 'Causas', 'Urbano/Rural'], axis=1)
x
```

```python
#Aqui, se importa, desde sklearn, un metodo de cross validation llamado train_test_split, que permitira
#dividir el conjunto de datos en 2, siendo uno para prueba y otro para entrenamiento
#De esta division, se crean 4 variables, donde 2 son para entrenamiento y 2 para prueba, donde las x son la tabla en si y las
#y son la caracteristica elegida, siendo en este caso CausasNumber.
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, df['CausasNumber'],
                                                    test_size=0.3, random_state=51
                                                    )
```

```python
#En esta linea, solamente se instancia la variable creada en el anterior grupo de codigo, permitiendo asi
#revisar lo que contiene esta variable
x_train
```

```python
#Se importa el algoritmo de RandomForest, a partir de sklearn, donde a partir de aqui, se realizaran las predicciones
#que utiliza el algoritmo
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=0)
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

```python
#En este ultimo grupo de codigo, se importan las metricas de accuracy, precision, recall y f1, donde, a traves de un print
#se mostraran en la terminal, los resultados que arrojaron este algoritmo
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print(f"acc: {accuracy_score(y_test,y_pred)}")
print(f"precision: {precision_score(y_test,y_pred,average='macro')}")
print(f"recall: {recall_score(y_test,y_pred,average='macro')}")
print(f"f1: {f1_score(y_test,y_pred,average='micro')}")
```
