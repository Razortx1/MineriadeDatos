```python
import pandas as pd
```

```python
df = pd.read_excel("/content/Conjunto de datos Accidente de transito Numero Pequeño.xlsx", sheet_name='default_1')
#df = pd.read_excel("/content/Conjunto de datos normalizado MaxMin.xlsx", sheet_name='default_1')
df
```

```python
x = df.copy()
x = x.drop(['Edad','Idaccidente' ,'Fecha', 'Hora', 'Mes', 'Comuna', 'Región', 'Calidad', 'Sexo', 'Resultado',
            'Accdtes.', 'Causas', 'Urbano/Rural'], axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, df['Edad'],
                                                    test_size=0.2, random_state=37
                                                    )
```

```python
x_train
```

```python
from sklearn.naive_bayes import CategoricalNB
```

```python
modelo = CategoricalNB(force_alpha=True)
```

```python
modelo.fit(x_train, y_train)
```

```python
y_pred = modelo.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, multilabel_confusion_matrix, ConfusionMatrixDisplay, confusion_matrix
confusion = multilabel_confusion_matrix(y_test,y_pred)
print(f"acc: {accuracy_score(y_test,y_pred)}")
print(f"precision: {precision_score(y_test,y_pred,average='macro')}")
print(f"recall: {recall_score(y_test,y_pred,average='macro')}")
print(f"f1: {f1_score(y_test,y_pred,average='micro')}")
```
